```python
# ファイル名をユーザーに入力してもらう
filename = input("ファイル名を入力してください（例：○○○.csv）: ")

# CSVファイルを読み込む
col_name = range(1, 12, 1)
df = pd.read_csv(filename, names=col_name, encoding='shift_jis')
df
```

```python
# 特定のセルの値を代入する：後に使用する
cell_j34 = df.at[33, 10]
cell_j34
```

```python
# 特定のセルの値を代入する：後に使用する
cell_j36 = df.at[35, 10]
cell_j36
```

```python
# 特定のセルの値を代入する：後に使用する
cell_j37 = df.at[36, 10]
cell_j37
```

```python
# データ分析に不要となる40行目までを削除
df.drop(range(0, 40), inplace=True)
df
```

```python
# カラム名を変更：分かりやすくする
df.rename(columns={1: 'No.', 10: 'トルク[%]', 11: '角度[°]'}, inplace=True)
df
```

```python
# No.、トルク、角度カラムのみのDataFrameを作成
peak_torque_df = df[['No.', 'トルク[%]', '角度[°]']].reset_index(drop=True)
peak_torque_df
```

```python
# 型の確認
peak_torque_df.dtypes
```

```python
# float型に変換
peak_torque_df[['トルク[%]', '角度[°]']] = peak_torque_df[['トルク[%]', '角度[°]']].astype(float)
```

```python
# DataFrame内の全てのセルを検索し、"EOF"が含まれる行を削除する
# 文字列型の"EOF"がNo.カラムにあるため、int型に変換できなかったため、"EOF"が含まれる行を削除するコードが必要となる
peak_torque_df = peak_torque_df[~peak_torque_df.apply(lambda row: row.astype(str).str.contains('EOF')).any(axis=1)]
peak_torque_df
```

```python
# int型に変換
peak_torque_df['No.'] = peak_torque_df['No.'].astype(int)
```

```python
# 型の確認
peak_torque_df.dtypes
```

```python
# トルクカラムの値の正負を変換する
peak_torque_df['トルク[%]'] = peak_torque_df['トルク[%]'] * -1
peak_torque_df
```

```python
# 要約統計量の確認：特にトルクのMAX値を把握する
peak_torque_df.describe()
```

```python
# トルク[%]のMAX値のインデックスを取得する：No.、トルク[%]、角度[°]カラムの情報を持っている
max_torque_row = peak_torque_df.loc[peak_torque_df['トルク[%]'].idxmax()]

# トルク[%]がMAXである行から任意の範囲を設定
min_index = max_torque_row.name - 510
max_index = max_torque_row.name + 100

# 範囲内のデータを抽出
selected_rows = peak_torque_df.loc[min_index:max_index, ['No.', '角度[°]', 'トルク[%]']]
```

```python
# selected_rowsの上から30行を表示：角度[°]が360°→0°になるNo.カラムを確認する
selected_rows.head(30)
```

```python
# 30行の中から角度カラムが最も小さい（最も0°に近い）インデックスを取得
min_index_selected_rows = selected_rows.head(30).loc[selected_rows['角度[°]'].idxmin()]
min_index_selected_rows
```

```python
# min_index_selected_rowsからNo.カラムの値を取得
min_index_selected_rows_no = min_index_selected_rows['No.']
min_index_selected_rows_no
```

```python
# selected_rowsの下から30行を表示：角度が0°→360°になるNo.カラムを確認する
selected_rows.tail(30)
```

```python
# 30行の中から角度カラムが最も大きい（最も360°に近い）インデックスを取得
max_index_selected_rows = selected_rows.tail(30).loc[selected_rows['角度[°]'].idxmax()]
max_index_selected_rows
```

```python
# max_index_selected_rowsからNo.カラムの値を取得
max_index_selected_rows_no = max_index_selected_rows['No.']
max_index_selected_rows_no
```

```python
# トルクがMAXである行を含む、角度0°～360°範囲を設定
min_index_selected_rows_index = min_index_selected_rows.name
max_index_selected_rows_index = max_index_selected_rows.name

# 角度0°～360°範囲内のデータを抽出（1サイクル分のデータ抽出）
selected_rows_0_360 = selected_rows.loc[min_index_selected_rows_index:max_index_selected_rows_index, ['No.', '角度[°]', 'トルク[%]']]
selected_rows_0_360
```

```python
# データ編集用に新たにDataFrame「selected_rows_0_360_Empty_column」を作成
selected_rows_0_360_Empty_column = selected_rows_0_360
selected_rows_0_360_Empty_column
```

```python
# selected_rows_0_360_Empty_columnに空のカラムを2列追加：後に変数の値を読み込むためのデータ領域が必要となる
selected_rows_0_360_Empty_column[['空カラム1', '空カラム2']] = None
selected_rows_0_360_Empty_column
```

```python
import os

# ファイルが存在するかチェック
if os.path.exists(f'{filename}_MAXトルク抽出.csv'):
    # ファイルが存在する場合の処理（上書き）
    # encoding無し、encoding='utf-8'において文字化けがあったため、encoding='shift_jis'とする。
    selected_rows_0_360.to_csv(f'{filename}_MAXトルク抽出.csv', index=False, encoding='shift_jis', mode='w')
else:
    # ファイルが存在しない場合の処理（新規作成）
    # encoding無し、encoding='utf-8'において文字化けがあったため、encoding='shift_jis'とする。
    selected_rows_0_360.to_csv(f'{filename}_MAXトルク抽出.csv', index=False, encoding='shift_jis')
```

```python
import csv

# CSVファイルを読み込む
with open(f'{filename}_MAXトルク抽出.csv', 'r', encoding='shift_jis') as f:
    reader = csv.reader(f)
    lines = list(reader)

# 特定のセルに変数の値を上書きする
lines[1][4] = cell_j34
lines[2][4] = cell_j35
lines[3][4] = cell_j36
lines[4][4] = cell_j37
lines[5][4] = min_index_selected_rows_no
lines[6][4] = max_index_selected_rows_no

# 更新した内容をCSVファイルに書き込む
with open(f'{filename}_MAXトルク抽出.csv', 'w', encoding='shift_jis', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(lines)
```
